## Inline Question 1: 
Did you notice anything about the comparative difficulty of training the three-layer network vs. training the five-layer network? In particular, based on your experience, which network seemed more sensitive to the initialization scale? Why do you think that is the case?

## Answer:
Training a deeper network like a five-layer network is generally more difficult than training a shallower one like a three-layer network. This is because deeper networks are more prone to problems like vanishing or exploding gradients, which can make the network hard to train effectively.

The five-layer network is likely more sensitive to the initialization scale. This is because in deeper networks, small changes in the weights can have a larger impact on the final output due to the multiple layers of computation. Therefore, if the weights are not initialized properly, it can lead to issues like vanishing or exploding gradients, making the network harder to train.

In contrast, a shallower network like a three-layer network has fewer layers of computation, so it's less likely to be affected by these issues and is generally less sensitive to the initialization scale.

## Inline Question 2:

AdaGrad, like Adam, is a per-parameter optimization method that uses the following update rule:

```
cache += dw**2
w += - learning_rate * dw / (np.sqrt(cache) + eps)
```

John notices that when he was training a network with AdaGrad that the updates became very small, and that his network was learning slowly. Using your knowledge of the AdaGrad update rule, why do you think the updates would become very small? Would Adam have the same issue?


## Answer: 
In AdaGrad, the cache variable accumulates the square of the gradients from all past steps. This means that if the gradients are consistently large, the cache value will grow quickly, causing the learning rate to effectively shrink. This can lead to very small updates and slow learning.

Adam, on the other hand, also maintains a running average of past gradients like AdaGrad, but it includes a "decay" factor that gives more weight to recent gradients. This means that if the gradients start to become smaller, the cache value in Adam will decrease faster than in AdaGrad, preventing the learning rate from shrinking too much. Therefore, Adam is less likely to have the same issue of updates becoming very small and learning slowing down.
